### Purchasing Items From The Inventory

```python
import json
```

```python
#reading from the 'records.json' which contains the inventory
fd = open('records.json', 'r')
r = fd.read()
fd.close()
inventory = json.loads(r)
```

```python
#displaying the inventory to the user
print("********** FOLLOWING PRODUCTS ARE AVAILABLE IN THE INVENTORY **********\n")
print("\n")
print (json.dumps(inventory, indent=3, default=str, ))
```

```python
#variable to keep track of each purchase
unique_id = 0
#dictionary to keep track of sales
sales = {}
```

```python
in_prodid = input("Enter the product id: ")
#check if id exists in the inventory
if in_prodid in inventory.keys():
  print("PRODUCT DETAILS: \n")
  #print the corrosponding product details
  print(inventory[in_prodid])

  in_qn = int(input("Enter the quantity: "))
  in_color = input("Enter color of your choice: ")
  #check if product for the input color exists
  if in_color in inventory[in_prodid]['prod_color']:
    #check if product for the input quantity exists
    if in_qn <= inventory[in_prodid]['prod_qn']:
      user_name = input("Enter your name: ")
      user_email = input("Enter your E-mail id: ")
      #decrement the value of quantity and remove the input colors from the inventory to confirm the purchase 
      inventory[in_prodid]['prod_qn'] = inventory[in_prodid]['prod_qn'] - in_qn
      inventory[in_prodid]['prod_color'].remove(in_color)
      #remove the product from inventory if product is sold out i.e 'prod_qn' = 0
      if inventory[in_prodid]['prod_qn'] == 0:
        del inventory[in_prodid]
      print("**************** Thank you for your purchase! **************** \n")
      #calculte the total amount for the purchase
      amount = inventory[in_prodid]['prod_price'] * in_qn
      #increment 'unique_id' everytime a purchase is made to keep a track of every purchase in the 'sales' dictionary
      unique_id += 1
      order_detail = f"\n Product ID: {in_prodid} \n Product Name: {inventory[in_prodid]['prod_name']} \n Product Color: {in_color} \n Quantity Ordered: {in_qn} \n Amount: {amount} \n"
      user_detail = f"\n Name: {user_name} \n E-mail id: {user_email} \n Your Unique ID: {unique_id} \n"

      print("ORDER DETAILS: ", order_detail)
      print("USER DETAILS: ", user_detail)

      #appending sales to the 'sales' dictionary
      sales[unique_id] = {'user_email': user_email, 'user_name': user_name, 'prod_id': in_prodid, 'amount': amount}
    
    else:
      print("Quantity insufficient, Try again!")
  else:
    print("That color is currently unavailable :( ")
else: 
  print("That product does not exist. Please try again!")
```

```python
#display the updated inventory after purchase 
print(json.dumps(inventory, indent=3, default=str, ))
```

```python
#display the sales
print(json.dumps(sales, indent=3, default=str, ))
```

```python
#write the updated inventory to the 'records.json' file 
js = json.dumps(inventory)
fd = open("records.json", 'w')
fd.write(js)
fd.close()
```

```python
#write the sales to 'sales.json' file
js = json.dumps(sales)
fd = open("sales.json", 'w')
fd.write(js)
fd.close()
```
